```python
from os import walk
import glob
```

```python
train_root = '../CASIA-WebFace' 
f = []
for (dirpath, dirnames, filenames) in walk(train_root):
    f.extend(dirnames)
    break
f
```

```python
#f = sorted(f)
print(f)
```

```python
print(len(f))
```

```python
new_f = f[:2000]
len(new_f)
```

```python
faces = {}
total_li = []
opt_fs = []
classes = 0
for dname in new_f:
    li = glob.glob(train_root + "/"+dname +"/*.jpg") 
    li_sd =  sorted(li)
    new_li = []
    for j in li_sd:
        new_li.append(j.split('WebFace/')[-1])
    faces[dname] = new_li
    total_li.extend(new_li)
    for i in new_li:
        opt_fs.append(i+'\t'+str(classes))
                     
    classes += 1
 
```
